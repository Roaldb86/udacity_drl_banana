## Algorithm

I chose to build a DQN and DDQN (Double DQN) to solve this task because these algorithms tend to to well when the state space is continous and the action space is discrete

The DQN algorthim uses a neural network to predict the Q_values instead of maintaining a table with observed values. This makes it possible to work with large state spaces as it generelizes well.

When the DQN agent collects observations, they are highly correlated as they are sequential and each state prime is allways a reult of the action taken in the last step. To mitigate this problem, It saves observations to a replay buffer and will on certain time steps sample randomly from this buffer to train its neural network, this is called experience replay.

Another issue with DQN is that its trying to predict both parts of the TD error with a neural network which is constantly evolving. This is like chasing a moving target and can easily lead to unstable network training. To mitigate this problem, it will use a seperate target network to predict the TD target value. This network is copied from the primary network on certain time intervals, but with a hyper parameter tau to control its convergence to the primary network. This leads to more stable trainig and is called fixed Q targets.

![TD_error](img/td_error.png)

The double DQN also uses the primary network to select actions instead of the max, and the target network to compute its Q value. This de-coupling of action selection/ value prediction increases performance of the algoritmn as can be shown in the results section of this report 

### Neural network

The network used in my solution is comprised of two hidden layers, the first with 64 nodes and the second with 128. the input layer is the state space with 37 nodes and the output layer has 4 nodes, 1 for each action. I have used ReLU activation for the hidden layers.
